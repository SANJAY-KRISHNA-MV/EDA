```python
correlation_data = df.select_dtypes(include=['float']).corr()
```

```python
plt.subplots(figsize=(10, 8))
plt.title('Correlation Heatmap',fontsize=20)
sns.heatmap(correlation_data, annot=True, cmap='coolwarm',center=0, linewidths=0.5, linecolor='black')
plt.show()
```

```python
float_cols = df.select_dtypes(include=['float']).columns
```

```python
df.groupby('Continent')[float_cols].mean().sort_values(by='World Population Percentage', ascending=False)
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df2 = df.groupby('Continent')[float_cols].mean().sort_values(by='2022 Population', ascending=False)
```

```python
df2
```

```python
df2.plot()
plt.show()
```

```python
df3 = df2.transpose()
```

```python
df3.plot()
plt.xticks(rotation=45)
plt.title('Mean Population by Continent', fontsize=20)
plt.show()
```
